```python
import re

dialogs_clean = []

for i in dialogs:
    # keep letters, apostrophe, and spaces
    i = re.sub("[^a-zA-Z' ]", "", i)
    i = i.lower().strip()
    dialogs_clean.append(i)

# join and split into words
all_words = " ".join(dialogs_clean).split()

# build frequency dict
words_dict = {}

for word in all_words:
    words_dict[word] = words_dict.get(word, 0) + 1

print(words_dict)
```

```python
dataset=pd.DataFrame()
dataset['Sentences']=dialogs_clean
dataset.head(20)
```

```python
#to get tokens
dataset['Sentences'][0].split()
```

```python
def create_unigram(sentence):
  tokens=sentence.split()
  unigram_list=[]
  for i in range(len(tokens)):
    unigram_list.append(tokens[i:i+1])
  return unigram_list
```

```python
def create_bigram(sentence):
  tokens=sentence.split()
  bigram_list=[]
  for i in range(len(tokens)-1):
    bigram_list.append(tokens[i:i+2])
  return bigram_list
```

```python
def create_trigram(sentence):
  token=sentence.split()
  trigram_list=[]

  for i in range(len(token)-2):
    trigram_list.append(token[i:i+3])
  return trigram_list
```

```python
final_unigram=[]
for i in range (dataset.shape[0]):
  final_unigram.append(create_unigram(dataset['Sentences'][i]))
dataset['unigram']=final_unigram
```

```python
final_bigram=[]
for i in range (dataset.shape[0]):
  final_bigram.append(create_bigram(dataset['Sentences'][i]))
dataset['bigram']=final_bigram
```

```python
final_trigram=[]
for i in range (dataset.shape[0]):
  final_trigram.append(create_trigram(dataset['Sentences'][i]))
dataset['trigram']=final_trigram
```

```python
from collections import Counter,defaultdict

model=defaultdict(lambda:defaultdict(lambda:0))

for i in range(dataset.shape[0]):
  for w1,w2,w3 in create_trigram(dataset['Sentences'][i]):
    model[(w1,w2)][w3]+=1
```

```python
unigram_dict={}
for i in tqdm(range(dataset.shape[0])):
  for word in dataset['unigram'][i]:
    if word[0] in unigram_dict:
      unigram_dict[word[0]]=unigram_dict[word[0]]+1
    else:
      unigram_dict[word[0]]=1
```

```python
counts=Counter(unigram_dict)
counts
```

```python
total_count=len(unigram_dict)
total_count
```

```python
for word in counts:
  counts[word]/=float(total_count)
counts
```

```python
for w1,w2 in model:
  total_count=float(sum(model[w1,w2].values()))
  for w3 in model[w1,w2]:
    model[(w1,w2)][w3]/=total_count
```

```python
import re

def predict_next_word(w1, w2):
    w1 = w1.lower()
    w2 = w2.lower()
    # Check if the bigram exists in the model
    if (w1, w2) in model:
        # Get the next word probabilities for the given bigram
        next_word_probs = model[(w1, w2)]
        # Find the word with the highest probability
        if next_word_probs:
            predicted_word = max(next_word_probs, key=next_word_probs.get)
            return predicted_word
    return None

# Example usage:
word1 = "want"
word2 = "to"
prediction = predict_next_word(word1, word2)
print(f"Next word prediction for '{word1} {word2}': {prediction}")

word1 = "hello"
word2 = "how"
prediction = predict_next_word(word1, word2)
print(f"Next word prediction for '{word1} {word2}': {prediction}")

word1 = "good"
word2 = "to"
prediction = predict_next_word(word1, word2)
print(f"Next word prediction for '{word1} {word2}': {prediction}")

word1 = "not"
word2 = "a"
prediction = predict_next_word(word1, word2)
print(f"Next word prediction for '{word1} {word2}': {prediction}")
```
